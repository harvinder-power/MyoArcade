```python
#Adds an index column at the start to normalise the timestamp parameter (NB only run once)
df.time = df.index
df.head()
```

```python
#low-pass filter using a Buterworth filter
import scipy.signal as signal

# First, design the Buterworth filter
N  = 2    # Filter order
Wn = 0.3 # Cutoff frequency (between 0-1, higher = less filtering)
B, A = signal.butter(N, Wn, output='ba')

emg1_f = signal.filtfilt(B,A, df.emg1)

fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(211)
plt.plot(df.index,df.emg1, 'b-')
plt.plot(df.index,emg1_f, 'r-',linewidth=2)
plt.ylabel("Signal intensity")
plt.xlabel("Time(s)")
plt.legend(['Original','Filtered'])
plt.title("Non-filtered vs filtered")
ax1.axes.get_xaxis().set_visible(False)
```
